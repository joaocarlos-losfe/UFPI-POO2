```python
from skimage.io import imread
from matplotlib.pyplot import imshow
from datetime import datetime

class Pessoa:
    
    __slots__ = ['_nome', '_cpf', '_endereco', '_telefone']
    
    def __init__(self, nome:str, cpf:str, endereco:str, telefone:str):
        self._nome = nome
        self._cpf = cpf
        self._endereco = endereco
        self._telefone = telefone 
    
    @property
    def nome(self) -> str:
        return self._nome
    
    @property
    def cpf(self) -> str:
        return self._cpf
    
    @property
    def endereco(self) -> str:
        return self._endereco
    
    @property
    def telefone(self) -> str:
        return self._telefone
        
class Fotografia:
        
    _quantidade_fotos = 0
    
    __slots__ = ['_foto_path', '_fotografo', '_data', '_proprietario']
    
    def __init__(self, foto_path:str, fotografo:Pessoa, Proprietario: Pessoa):
        
        self._foto_path = foto_path
        self._fotografo = fotografo
        self._data = datetime.now().strftime("%d/%m/%Y") + " " + datetime.now().strftime("%H:%M:%S")
        self._proprietario = proprietario
        Fotografia._quantidade_fotos +=1
    
    @property
    def foto_path(self) -> str:
        return self._foto_path
    
    @foto_path.setter
    def foto_path(self, path:str):
        self._foto_path = path
        
    def mostrar_fotografia(self):
        img = imread(self._foto_path)
        imshow(img)
    
    def tamanho_fotografia(self) -> str:
        img = imread(self._foto_path)
        return f"resolução da imagem: {img.shape[1]} x {img.shape[0]} pixels"
    
    @property
    def fotografo(self) -> Pessoa:
        return self._fotografo
    
    @property
    def proprietario(self) -> Pessoa:
        return self._proprietario
    
    @property
    def data(self) -> str:
        return self._data
    
    @staticmethod
    def total_fotos() -> int:
        return Fotografia._quantidade_fotos

fotografo = Pessoa("marcos", "111.111.111-11", "picos - PI", "(99)99988-9999")
proprietario = Pessoa("jose marcos", "222.222.222", "Valença", "(88)88988-8999")
foto1 = Fotografia("./img/img1.jpg", fotografo, proprietario)

fotografo = Pessoa("elias", "333.333.333-33", "Teresina - PI", "(55)59555-9999")
foto2 = Fotografia("./img/space.png", fotografo, proprietario)
             
```
